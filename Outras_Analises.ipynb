#IMPORTANDO BIBLIOTECAS

```python
import pandas as pd
import plotly.express as plt
```

#IMPORTANDO DADOS 

```python
tabela = pd.read_csv('/content/Base_Teste_Estagio.xls',encoding='latin1')
display(tabela)
```

#REMOVENDO TODAS AS COLUNAS QUE POSSUEM ALGUM VALOR VAZIO

```python
tabela = tabela.dropna()
display(tabela.info())
display(tabela.describe())
```

```python
qt_sistemas = tabela['Sistema_Operacional'].value_counts()
display(qt_sistemas)

qt_sistemas.perc = tabela['Sistema_Operacional'].value_counts(normalize=True)
display(qt_sistemas.perc.round(1))
```

#REMOVENDO COLUNA ID

```python
tabela = tabela.drop('ID', axis=1)
display(tabela)
```

#CRIANDO HISTROGRAMA PARA CADA COLUNA DA TABELA

```python
for colunas in tabela:
  grafico = plt.histogram(tabela, x=colunas, color='Sistema_Operacional')
  grafico.show()
```
